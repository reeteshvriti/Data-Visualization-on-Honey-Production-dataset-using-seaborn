### 3. Display the percentage distribution of the data in each year using the pie chart.

```python
import matplotlib.pyplot as plt


# Calculate the percentage distribution of data in each year
years = df_data['year'].value_counts(normalize=True) * 100

# Create a pie chart of the percentage distribution
plt.pie(years, labels=years.index.astype(str), autopct='%1.1f%%')
plt.title('Percentage Distribution of Data by Year')
plt.show()


# The percentages of the data by year is more 2003, 2002 and 2001
```

### 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python

sns.displot(data=df_data, x='priceperlb')

# the highest count is 100
```

### 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
# Create a scatter plot of numcol vs. prodval
plt.scatter(df_data['numcol'], df_data['prodvalue'])
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value ($)')
plt.title('Number of Colonies vs. Production Value')
plt.show()

# Number of Colonies and Production Value is linearly distributed
```

### 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.


```python
sns.boxplot(data=df_data, x='year', y='prodvalue')
plt.xlabel('Year')
plt.ylabel('Production Value ($)')
plt.title('Production Value by Year')
plt.show()

# The Production Value is more in the year 2010
```

### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python

cols_of_interest = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']
df_filtered = df_data[cols_of_interest]

# Creating pairplot
sns.pairplot(data=df_filtered, hue='year')
plt.show()

# The below pair plot shows the distribution of all values
```

### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
cols_of_interest = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
df_filtered = df_data[cols_of_interest]

corr = df_filtered.corr()

# Creating heatmap of correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# numcol and totalprod has strong releationship
```
